# Gradient Descent

Sometimes called the steepest descent. It is an opimisation algorithm to find the minimum of a function

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

## Example 1 - Cost functions

$f(x) = x^2 + x + 1$

```python
def f(x):
    return x**2 + x + 1
```

```python
# generate the data 
x1 = np.linspace(start= -3, stop=3, num=100)
x1
```

```python
plt.style.use('dark_background')
plt.figure(figsize=(9,6))
plt.plot(x1, f(x1))
plt.xlim([-3.0, 3.0])
plt.ylim(0, 15)
plt.xlabel('x value', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.grid(color='w', alpha=0.15);
```

## Slope & Derivatives

> Create the derivative of `f(x)` function above

The calculus power rule. 

$d/dx xn = n * xn-1$


```python
def df(x):
    return 2 * x + 1
```

```python
# plot function and derivative side by side
# to layout subplots the subplot(x,y,z) is used. 
# where x = rows, y = columns, z = position/place. 
# eg. (1,2,1) equals 1 row of 2 columns and position 1. 

plt.style.use('dark_background')
plt.figure(figsize=(18,6))

# cost function 
plt.subplot(1, 2, 1)
plt.plot(x1, f(x1))
plt.xlim([-3.0, 3.0])
plt.ylim(0, 15)
plt.xlabel('x value', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.grid(color='w', alpha=0.15)

# derivative 
plt.subplot(1,2,2)
plt.plot(x1, df(x1))
plt.xlim([-3.0, 3.0])
plt.ylim(-6, 6)
plt.xlabel('slope of cost function', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.grid(color='w', alpha=0.15);
```

## Create our Gradient Descent Algorithm

```python
new_x = 3
prev_x = 0
step_multiplier = 0.1
precision = 0.0001

x_list = [new_x]
slopes = [df(new_x)]

for n in range(500):
    prev_x = new_x
    gradient = df(prev_x)
    new_x = prev_x - step_multiplier * gradient
    
    step_size = abs(new_x - prev_x)
    x_list.append(new_x)
    slopes.append(df(new_x))
    
    if step_size < precision:
        print('Iterations: ', n)
        break

print('local min occurs at: ', new_x)
print('Slope of df(x):', df(new_x))
print('f(x) value is:', f(new_x))
    
```

```python
# add the gradient descent data

plt.style.use('dark_background')
plt.figure(figsize=(20,5))

# cost function 
plt.subplot(1, 3, 1)
plt.plot(x1, f(x1), alpha=0.6)
plt.xlim([-3.0, 3.0])
plt.ylim(0, 15)
plt.xlabel('x value', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.grid(color='w', alpha=0.15)

# add scatter plot of x values
xvals = np.array(x_list)
plt.scatter(x_list, f(xvals), color='#e449f2', s=70, alpha=0.7)

# derivative 
plt.subplot(1,3,2)
plt.plot(x1, df(x1), alpha=0.6)
plt.xlim([-3.0, 3.0])
plt.ylim(-5, 8)
plt.xlabel('slope of cost function', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.grid(color='w', alpha=0.15)

plt.scatter(x_list, slopes, color='#e449f2', s=70, alpha=0.7);

# derivative close up 
plt.subplot(1,3,3)
plt.plot(x1, df(x1), alpha=0.6)
plt.xlim([-0.55, -0.4])
plt.ylim(-0.03, 0.8)
plt.xlabel('slope (close up)', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.grid(color='w', alpha=0.15)

plt.scatter(x_list, slopes, color='#e449f2', s=70, alpha=0.7);
```

## Example 2 - Multiple minima vs Initial guess

$g(x) = x^4 - 4x^2 + 5$

```python
# make our data 
x2 = np.linspace(-2, 2, 1000)
```

```python
def g(x):
    return x**4 - 4*x**2 + 5
```

```python
def dg(x):
    return 4*x**3 - 8*x
```

```python
# plot 

plt.style.use('dark_background')
plt.figure(figsize=(20,5))

# cost function 
plt.subplot(1, 2, 1)
plt.xlim([-2.0, 2.0])
plt.ylim(0.5, 5.5)
plt.xlabel('x value', fontsize=16)
plt.ylabel('g(x)', fontsize=16)
plt.grid(color='w', alpha=0.15)
plt.plot(x2, g(x2))


# derivative 
plt.subplot(1,2,2)
plt.xlim([-2.0, 2.0])
plt.ylim(-6, 8)
plt.xlabel('slope of cost function', fontsize=16)
plt.ylabel('dg(x)', fontsize=16)
plt.grid(color='w', alpha=0.15)
plt.plot(x2, dg(x2));
```

### Gradient descent algorithm

Create a python function to cover the algorithm

```python
def gradient_descent(derivatve_func, initial_guess, multiplier = 0.02, precision = 0.001, max_iter=300):
    '''
    derivative_func: accepts a passed derivative function.
    initial_guess: We supply an initial guess which is set as an X value. 
    multiplier: 
    precision: 
    '''
    new_x = initial_guess

    x_list = [new_x]
    slopes = [derivatve_func(new_x)]

    for n in range(max_iter):
        prev_x = new_x
        gradient = derivatve_func(prev_x)
        new_x = prev_x - multiplier * gradient

        step_size = abs(new_x - prev_x)
        x_list.append(new_x)
        slopes.append(derivatve_func(new_x))

        if step_size < precision:
            break
    return new_x, x_list, slopes
```

```python
local_min, list_x, deriv_list = gradient_descent(dg, 0.5, 0.02, 0.001)
print('local_min occurs at: ', local_min)
print('Number of steps: ', len(list_x))
```

```python
# using named parameters 
local_min, list_x, deriv_list = gradient_descent(derivatve_func=dg, initial_guess=-0.5, multiplier=0.02, precision=0.001)
print('local_min occurs at: ', local_min)
print('Number of steps: ', len(list_x))
```

```python
# omiting the params wit defaults 
local_min, list_x, deriv_list = gradient_descent(derivatve_func=dg, initial_guess=-0.2)
print('local_min occurs at: ', local_min)
print('Number of steps: ', len(list_x))
```

```python
# plots gradient descent function (plot function, derivitive and scatter)
local_min, list_x, deriv_list = gradient_descent(derivatve_func=dg, initial_guess=0.1)

plt.style.use('dark_background')
plt.figure(figsize=(20,5))

# cost function 
plt.subplot(1, 2, 1)
plt.xlim([-2.0, 2.0])
plt.ylim(0.5, 5.5)
plt.xlabel('x value', fontsize=16)
plt.ylabel('g(x)', fontsize=16)
plt.grid(color='w', alpha=0.15)
plt.plot(x2, g(x2))

plt.scatter(list_x, g(np.array(list_x)), color='#e449f2', alpha=0.7)


# derivative 
plt.subplot(1,2,2)
plt.xlim([-2.0, 2.0])
plt.ylim(-6, 8)
plt.xlabel('slope of cost function', fontsize=16)
plt.ylabel('dg(x)', fontsize=16)
plt.grid(color='w', alpha=0.15)
plt.plot(x2, dg(x2))

plt.scatter(list_x, deriv_list, color='#e449f2', alpha=0.7);
```

## Example 3 - Divergence and Overflow
#### Using python tuples 

## $h(x) = x^5 - 2x^4 + 2$

```python
# make our data 
x3 = np.linspace(-2.5, 2.5, 1000)
```

```python
def h(x):
    return x**5 - 2*x**4 + 2
```

```python
def dh(x):
    return 5*x**4 - 8*x**3
```

```python
# plots gradient descent function (plot function, derivitive and scatter)
local_min, list_x, deriv_list = gradient_descent(derivatve_func=dh, initial_guess=0.2)

plt.style.use('dark_background')
plt.figure(figsize=(20,5))

# cost function 
plt.subplot(1, 2, 1)
plt.xlim([-1.2, 2.5])
plt.ylim(-1, 2.5)
plt.xlabel('x value', fontsize=16)
plt.ylabel('h(x)', fontsize=16)
plt.grid(color='w', alpha=0.15)
plt.plot(x3, h(x3))

plt.scatter(list_x, h(np.array(list_x)), color='#e449f2', alpha=0.7)


# derivative 
plt.subplot(1,2,2)
plt.xlim([-1.0, 2.0])
plt.ylim(-4, 2.5)
plt.xlabel('slope of cost function', fontsize=16)
plt.ylabel('dh(x)', fontsize=16)
plt.grid(color='w', alpha=0.15)
plt.plot(x3, dh(x3))

plt.scatter(list_x, deriv_list, color='#e449f2', alpha=0.7);
```

```python
print('local_min occurs at: ', local_min)
print('cost at minimum: ', h(local_min))
print('Number of steps: ', len(list_x))
```

```python
# plots gradient descent function 
# (plot function, derivitive and scatter)
# use a left hand side plot 
local_min, list_x, deriv_list = gradient_descent(derivatve_func=dh, initial_guess=-0.2, max_iter=70)

plt.style.use('dark_background')
plt.figure(figsize=(20,5))

# cost function 
plt.subplot(1, 2, 1)
plt.xlim([-1.2, 2.5])
plt.ylim(-1, 2.5)
plt.xlabel('x value', fontsize=16)
plt.ylabel('h(x)', fontsize=16)
plt.grid(color='w', alpha=0.15)
plt.plot(x3, h(x3))

plt.scatter(list_x, h(np.array(list_x)), color='#e449f2', alpha=0.7)


# derivative 
plt.subplot(1,2,2)
plt.xlim([-1.0, 2.0])
plt.ylim(-4, 2.5)
plt.xlabel('slope of cost function', fontsize=16)
plt.ylabel('dh(x)', fontsize=16)
plt.grid(color='w', alpha=0.15)
plt.plot(x3, dh(x3))

plt.scatter(list_x, deriv_list, color='#e449f2', alpha=0.7);
```

```python
print('local_min occurs at: ', local_min)
print('cost at minimum: ', h(local_min))
print('Number of steps: ', len(list_x))
```

## The Learning rate 

The learning rate is the difference between the gradient and the multiplier and precision. The Issues faced here are between settings that are too high and the algorithm never converging, or the buffer overflow infinite error and the settings being too tight and it taking a very long time to converge. 

```python
# plots gradient descent function (plot function, derivitive and scatter)
n = 100
low_gamma = gradient_descent(derivatve_func=dg, initial_guess=3, multiplier=0.0005, precision=0.0001, max_iter=n)

mid_gamma = gradient_descent(derivatve_func=dg, initial_guess=3, multiplier=0.001, precision=0.0001, max_iter=n)

high_gamma = gradient_descent(derivatve_func=dg, initial_guess=3, multiplier=0.002, precision=0.0001, max_iter=n)

insane_gamma = gradient_descent(derivatve_func=dg, initial_guess=1.9, multiplier=0.25, precision=0.0001, max_iter=n)


plt.style.use('dark_background')
plt.figure(figsize=(20,6))

# cost function 
plt.subplot(1, 2, 1)
plt.xlim([0, n])
plt.ylim(0, 50)
plt.title('Effect of the learning rate')
plt.xlabel('n iterations', fontsize=16)
plt.ylabel('Cost', fontsize=16)
plt.grid(color='w', alpha=0.15)

# values for the charts
# Y axis data: convert lists to np arrays 
low_vals = np.array(low_gamma[1])
mid_vals = np.array(mid_gamma[1])
high_vals = np.array(high_gamma[1])
insanity = np.array(insane_gamma[1])

# X axis data: list from 0 to n+1
iter_list = list(range(0, n+1))

plt.plot(iter_list, g(low_vals));
plt.scatter(iter_list,g(low_vals), color='#e449f2', alpha=0.7)

plt.plot(iter_list, g(mid_vals));
plt.scatter(iter_list,g(mid_vals), color='steelblue', alpha=0.7)

plt.plot(iter_list, g(high_vals));
plt.scatter(iter_list,g(high_vals), color='b', alpha=0.7)

# cost function 
plt.subplot(1, 2, 2)
plt.xlim([0, n])
plt.ylim(0, 50)
plt.title('Effect of the learning rate (non convergance)')
plt.xlabel('n iterations', fontsize=16)
plt.ylabel('Cost', fontsize=16)
plt.grid(color="w", alpha=0.15)

# insanity 
plt.plot(iter_list, g(insanity));
plt.scatter(iter_list,g(insanity), color='#ff6090', alpha=0.7)
```
