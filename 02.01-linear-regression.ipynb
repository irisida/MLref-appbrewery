## Section 2 - Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression

%matplotlib inline
```

```python
data = pd.read_csv('./data/cost_revenue_clean.csv')
data
```

```python
data.describe()
```

```python
x = pd.DataFrame(data, columns=['budget_usd'])
y = pd.DataFrame(data, columns=['world_usd'])
```

```python
# plot the data
plt.style.use('dark_background')
plt.figure(figsize=(10,6))
plt.scatter(x, y, alpha=0.3)
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.grid(color='w', alpha=0.2)
plt.title("Movie cost and Global revenue")
plt.xlabel("Production budget USD")
plt.ylabel("Gross revenues USD");
```

```python
reg = LinearRegression()
reg.fit(x, y)
```

```python
# slope coefficient
reg.coef_
```

```python
# intercept
reg.intercept_
```

```python
# New plot of the data and the regression line. 
plt.style.use('dark_background')
plt.figure(figsize=(10,6))
plt.scatter(x, y, alpha=0.3)
plt.plot(x, reg.predict(x), color='#e449f2', linewidth=2, alpha=0.5)
plt.grid(color='w', alpha=0.2)
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.title("Movie cost and Global revenue")
plt.xlabel("Production budget USD")
plt.ylabel("Gross revenues USD");
```

```python
reg.score(x, y)
```
